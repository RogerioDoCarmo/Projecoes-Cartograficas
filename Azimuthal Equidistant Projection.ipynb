```python
import os 

#os.environ['PROJ_LIB'] = 'C:/ProgramData/Anaconda3/share/proj'
os.environ['PROJ_LIB'] = 'C:/ProgramData/Anaconda3/pkgs/proj4-5.2.0-ha925a31_1/Library/share'
desktop = os.path.join(os.path.join(os.environ['USERPROFILE']), 'Desktop') 

from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
#width = 5000000;
lon_0 = -10.79908; lat_0 = 27.04604
m = Basemap(llcrnrlat=-90,urcrnrlat=90,
            llcrnrlon=-180,urcrnrlon= 180, projection='aeqd',
            lat_0=lat_0,lon_0=lon_0)
# fill background.
m.drawmapboundary(fill_color='aqua')
# draw coasts and fill continents.
m.drawcoastlines(linewidth=0.5)
m.fillcontinents(color='coral',lake_color='aqua')
# 20 degree graticule.
m.drawparallels(np.arange(-80,80,10))
m.drawmeridians(np.arange(-180,180,10))
# draw a black dot at the center.
xpt, ypt = m(lon_0, lat_0)
m.plot([xpt],[ypt],'ko')
# draw the title.
plt.title('Azimuthal Equidistant Projection')
plt.draw()
plt.savefig(desktop + '\\Azimutal Equidistant Projection.png', bbox_inches='tight')
plt.show()
```
